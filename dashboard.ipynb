```python
# 📊 Дашборд карьерного роста (полная версия)
# Автоматически загружает данные из GitHub и строит графики

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import requests
import io
import numpy as np

# ========== НАСТРОЙКИ ==========
GITHUB_RAW_URL = "https://raw.githubusercontent.com/EkaterinaLavlinskaya/ai-career-log/main/data/interactions.jsonl"
# ================================

# Загружаем данные
print("🔄 Загружаем данные...")
response = requests.get(GITHUB_RAW_URL)
data = response.text

# Читаем в pandas
df = pd.read_json(io.StringIO(data), lines=True)
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Сортируем по времени
df = df.sort_values('timestamp')

print(f"✅ Загружено записей: {len(df)}")
print(f"📅 Период: {df['timestamp'].min().date()} — {df['timestamp'].max().date()}")

# ========== 1. ДИНАМИКА ПРОЕКТОВ ==========
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(df['timestamp'], df['projects'], marker='o', linestyle='-', color='#2E86AB')
plt.title('📈 Рост количества проектов')
plt.xlabel('Дата')
plt.ylabel('Проекты')
plt.grid(True, alpha=0.3)

# ========== 2. ДИНАМИКА ОТКЛИКОВ ==========
plt.subplot(1, 2, 2)
plt.plot(df['timestamp'], df['applications'], marker='o', linestyle='-', color='#A23B72')
plt.title('📬 Отклики на вакансии')
plt.xlabel('Дата')
plt.ylabel('Отклики')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# ========== 3. СРАВНЕНИЕ ПОКАЗАТЕЛЕЙ ==========
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(df['timestamp'], df['positive_responses'], marker='o', label='✅ Положительные ответы', color='#F18F01')
plt.plot(df['timestamp'], df['tests'], marker='o', label='🧪 Тесты', color='#C73E1D')
plt.title('📊 Успехи vs Тесты')
plt.xlabel('Дата')
plt.ylabel('Количество')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(df['timestamp'], df['interviews'].fillna(0), marker='o', label='🤝 Собеседования', color='#2E86AB')
plt.plot(df['timestamp'], df['offers'].fillna(0), marker='o', label='🎉 Офферы', color='#A23B72')
plt.title('🎯 Собеседования и офферы')
plt.xlabel('Дата')
plt.ylabel('Количество')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ========== 4. ВОРОНКА ОТКЛИКОВ ==========
last = df.iloc[-1]

stages = [
    "📬 Отклики",
    "📋 На рассмотрении",
    "🤝 Собеседования",
    "🎉 Офферы",
    "❌ Отказы",
    "🗑️ Удалено"
]

values = [
    last.get("applications", 0),
    last.get("review", 0),
    last.get("interviews", 0),
    last.get("offers", 0),
    last.get("rejections", 0),
    last.get("closed_vacancy", 0)
]

plt.figure(figsize=(10, 6))
bars = plt.barh(stages, values, color=['#2E86AB', '#F18F01', '#A23B72', '#C73E1D', '#E3845F', '#6A6A6A'])

for i, (bar, v) in enumerate(zip(bars, values)):
    plt.text(bar.get_width() + 0.5, bar.get_y() + bar.get_height()/2, str(v), va='center', fontsize=11)

plt.title("📊 Воронка откликов")
plt.xlabel("Количество")
plt.tight_layout()
plt.show()

# Конверсия
apps = values[0]
if apps > 0:
    print(f"\n📈 Конверсия отклик → рассмотрение: {values[1]/apps*100:.1f}%")
    print(f"📈 Конверсия рассмотрение → собеседование: {values[2]/values[1]*100:.1f}%" if values[1] > 0 else "📈 Конверсия рассмотрение → собеседование: —")
    print(f"📈 Конверсия собеседование → оффер: {values[3]/values[2]*100:.1f}%" if values[2] > 0 else "📈 Конверсия собеседование → оффер: —")
    print(f"🎯 Общая конверсия отклик → оффер: {values[3]/apps*100:.1f}%")

# ========== 5. ТЕКУЩАЯ СТАТИСТИКА ==========
print("\n" + "="*50)
print("📌 ТЕКУЩЕЕ СОСТОЯНИЕ")
print("="*50)
print(f"📅 Последнее обновление: {last['timestamp']}")
print(f"📁 Проектов: {last['projects']}")
print(f"🎓 Сертификатов: {last['certificates']}")
print(f"📬 Откликов: {last['applications']}")
print(f"📋 На рассмотрении: {last.get('review', 0)}")
print(f"🤝 Собеседований: {last.get('interviews', 0)}")
print(f"✅ Положительных ответов: {last['positive_responses']}")
print(f"🧪 Тестов: {last['tests']}")
print(f"🎉 Офферов: {last.get('offers', 0)}")
print(f"❌ Отказов: {last.get('rejections', 0)}")
print(f"🗑️ Вакансий удалено: {last.get('closed_vacancy', 0)}")
print(f"📝 Последняя заметка: {last['note']}")

# ========== 6. ПРОГНОЗ (опционально) ==========
if len(df) >= 3:
    x = np.arange(len(df)).reshape(-1, 1)
    y = df['projects'].values.reshape(-1, 1)
    from sklearn.linear_model import LinearRegression
    model = LinearRegression()
    model.fit(x, y)

    future_days = 30
    future_x = np.arange(len(df) + future_days).reshape(-1, 1)
    future_y = model.predict(future_x)

    plt.figure(figsize=(10, 4))
    plt.plot(df['timestamp'], df['projects'], 'o-', label='Факт')
    future_dates = pd.date_range(start=df['timestamp'].max() + pd.Timedelta(days=1), periods=future_days)
    plt.plot(future_dates, future_y[-future_days:], '--', label='Прогноз (30 дней)')
    plt.plot(df['timestamp'], df['applications'].fillna(0))
    plt.axhline(y=15, color='red', linestyle=':', label='Цель (15 проектов)')
    plt.title('🔮 Прогноз роста проектов')
    plt.xlabel('Дата')
    plt.ylabel('Проекты')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

    target = 15
    target_idx = np.where(future_y.flatten() >= target)[0]
    if len(target_idx) > 0:
        days_to_target = target_idx[0] - len(df) + 1
        print(f"\n🎯 При текущем темпе цель в {target} проектов будет достигнута через {days_to_target} дней")

```
