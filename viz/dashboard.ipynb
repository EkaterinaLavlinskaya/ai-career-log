```python
# 📊 Дашборд карьерного роста
# Автоматически загружает данные из GitHub и строит графики

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import requests
import io

# ========== НАСТРОЙКИ ==========
GITHUB_RAW_URL = "https://raw.githubusercontent.com/EkaterinaLavlinskaya/ai-career-log/main/data/interactions.jsonl"
# ================================

# Загружаем данные
print(" Загружаем данные...")
response = requests.get(GITHUB_RAW_URL)
data = response.text

# Читаем в pandas
df = pd.read_json(io.StringIO(data), lines=True)
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Сортируем по времени
df = df.sort_values('timestamp')

print(f" Загружено записей: {len(df)}")
print(f"Период: {df['timestamp'].min().date()} — {df['timestamp'].max().date()}")

# ========== 1. ДИНАМИКА ПРОЕКТОВ ==========
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(df['timestamp'], df['projects'], marker='o', linestyle='-', color='#2E86AB')
plt.title(' Рост количества проектов')
plt.xlabel('Дата')
plt.ylabel('Проекты')
plt.grid(True, alpha=0.3)

# ========== 2. ДИНАМИКА ОТКЛИКОВ ==========
plt.subplot(1, 2, 2)
plt.plot(df['timestamp'], df['applications'], marker='o', linestyle='-', color='#A23B72')
plt.title('📬 Отклики на вакансии')
plt.xlabel('Дата')
plt.ylabel('Отклики')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# ========== 3. СРАВНЕНИЕ ПОКАЗАТЕЛЕЙ ==========
plt.figure(figsize=(10, 5))
plt.plot(df['timestamp'], df['positive_responses'], marker='o', label=' Положительные ответы', color='#F18F01')
plt.plot(df['timestamp'], df['tests'], marker='o', label='🧪 Тестовые задания', color='#C73E1D')
plt.title(' Успехи vs Тесты')
plt.xlabel('Дата')
plt.ylabel('Количество')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# ========== 4. ТЕКУЩАЯ СТАТИСТИКА ==========
last = df.iloc[-1]
print("\n" + "="*50)
print(" ТЕКУЩЕЕ СОСТОЯНИЕ")
print("="*50)
print(f"Последнее обновление: {last['timestamp']}")
print(f" Проектов: {last['projects']}")
print(f" Сертификатов: {last['certificates']}")
print(f" Откликов: {last['applications']}")
print(f" Положительных ответов: {last['positive_responses']}")
print(f" Тестов: {last['tests']}")
print(f" Последняя заметка: {last['note']}")

# ========== 5. ДОПОЛНИТЕЛЬНО: ПРОГНОЗ (опционально) ==========

import numpy as np
if len(df) >= 3:
    # Простая линейная регрессия
    x = np.arange(len(df)).reshape(-1, 1)
    y = df['projects'].values.reshape(-1, 1)
    from sklearn.linear_model import LinearRegression
    model = LinearRegression()
    model.fit(x, y)

    future_days = 30
    future_x = np.arange(len(df) + future_days).reshape(-1, 1)
    future_y = model.predict(future_x)

    plt.figure(figsize=(10, 4))
    plt.plot(df['timestamp'], df['projects'], 'o-', label='Факт')
    future_dates = pd.date_range(start=df['timestamp'].max() + pd.Timedelta(days=1), periods=future_days)
    plt.plot(future_dates, future_y[-future_days:], '--', label='Прогноз (30 дней)')
    plt.axhline(y=15, color='red', linestyle=':', label='Цель (15 проектов)')
    plt.title('🔮 Прогноз роста проектов')
    plt.xlabel('Дата')
    plt.ylabel('Проекты')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

    # Когда достигнем 15 проектов?
    target = 15
    target_idx = np.where(future_y.flatten() >= target)[0]
    if len(target_idx) > 0:
        days_to_target = target_idx[0] - len(df) + 1
        print(f"\n При текущем темпе цель в {target} проектов будет достигнута через {days_to_target} дней")
# ========== ВОРОНКА ОТКЛИКОВ ==========
import pandas as pd
import matplotlib.pyplot as plt
import requests
import io

# Загружаем данные (как в дашборде)
GITHUB_RAW_URL = "https://raw.githubusercontent.com/EkaterinaLavlinskaya/ai-career-log/main/data/interactions.jsonl"
response = requests.get(GITHUB_RAW_URL)
df = pd.read_json(io.StringIO(response.text), lines=True)

# Берём последнюю запись
last = df.iloc[-1]

# Данные для воронки
stages = [
    " Отклики",
    " Тесты",
    "Собеседования",
    " Офферы"
]

values = [
    last.get("applications", 0),
    last.get("tests", 0),
    last.get("interviews", 0),
    last.get("offers", 0)
]

# Рисуем воронку
plt.figure(figsize=(10, 6))
plt.barh(stages, values, color=['#2E86AB', '#A23B72', '#F18F01', '#C73E1D'])

for i, v in enumerate(values):
    plt.text(v + 0.5, i, str(v), va='center', fontsize=12)

plt.title(" Воронка откликов → офферы")
plt.xlabel("Количество")
plt.tight_layout()
plt.show()

# Конверсия
if values[0] > 0:
    print(f"\n Конверсия отклик → тест: {values[1]/values[0]*100:.1f}%")
if values[1] > 0:
    print(f" Конверсия тест → собеседование: {values[2]/values[1]*100:.1f}%")
if values[2] > 0:
    print(f" Конверсия собеседование → оффер: {values[3]/values[2]*100:.1f}%")
if values[0] > 0:
    print(f" Общая конверсия отклик → оффер: {values[3]/values[0]*100:.1f}%")
```
